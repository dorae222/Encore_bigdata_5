# 데이터 적재 및 통합

```python
# 라이브러리 가져오기
import pandas as pd
```

```python
# 데이터 가져오기
data = pd.read_excel('data/NC Dinos.xlsx',sheet_name = None)
data
```

```python
# 시트별 데이터를 각각 개별 변수에 저장
nc13,nc14,nc15 = data.values()
```

```python
print('nc13.shape:',nc13.shape)
print('nc14.shape:',nc14.shape)
print('nc15.shape:',nc15.shape)
```

```python
nc13.columns
```

```python
nc15.head(2)
```

```python
# nc13,nc14,nc15의 Unnamed: 0컬럼 삭제
nc13.drop(['Unnamed: 0'],axis=1,inplace=True)
nc14.drop(['Unnamed: 0'],axis=1,inplace=True)
nc15.drop(['Unnamed: 0'],axis=1,inplace=True)
# del nc13['Unnamed: 0']
```

```python
nc13.head(2)
```

```python
# nc13, nc14, nc15 선수 자료중 '선수명' 컬럼을 이용해 선수명에 동명이인이 있는지 여부와
# 선수 명수를 출력하기' 
print('nc13:',len(nc13['선수명'].unique()),'\n',nc13['선수명'].unique())
print('nc14:',len(nc14['선수명'].unique()),'\n',nc14['선수명'].unique())
print('nc15:',len(nc15['선수명'].unique()),'\n',nc15['선수명'].unique())
```

```python
# set(집합 자료형)은 내부에 중복데이터 입력시 1개로 집계
# set()을 이용하면 다년간 주전이었던 선수 확인 가능
# 13~15년까지는 항상 20명의 선수가 데이터로 집계됨
# 14~~15년 연속 주전이었던 선수의 인원수
print(set(nc14['선수명'].unique()) & set(nc15['선수명'].unique()))
print(len(set(nc14['선수명'].unique()) & set(nc15['선수명'].unique())))
```

## 최종데이터

- 2013, 2014, 2015 데이터를 모두 합치되 각각이 몇년도 데이터인지 알 수 있도록 '연도'값을 추가

```python
# nc13,nc14,nc15 변수에 '연도'추가
nc13['연도'] = 2013
nc14['연도'] = 2014
nc15['연도'] = 2015
```

```python
nc13.head(2)
```

## 여러개의 데이터 프레임 결합

```python
# 결합 방식: 컬럼개수 및 속성이 완전히 같으므로 row길이를 늘려준다
# 최종결과는: 20+20+20 -> (60,20)
nc_result1=pd.concat([nc13,nc14,nc15],ignore_index=True)
nc_result1
```

```python
nc_result1.shape
```

```python
# 조건색인을 이용해 2013,2014년 데이터만 조회
#nc_result1[(nc_result1['연도'] == 2013) | (nc_result1['연도'] == 2014)]
nc_result1[(nc_result1['연도'] != 2015)]
```

## 2015년 타율이 높은 상위 10명
- 동일 타율을 기록한 선수에 대해서는 안타 개수가 많은 순서로 정렬

```python
#nc_result1[(nc_result1['연도'] == 2015)].sort_values(by=['타율','안타'],ascending=[False,False])
tmp = nc_result1[(nc_result1['연도'] == 2015)].sort_values(by=['타율','안타'],ascending=[False,False])
tmp.head(10)
```

```python
# 특정 컬럼에 .tolist()를 사용하면 파이썬 리스트로 뽑을 수도 있다
tmp['선수명'].tolist()
```

## 2013~2015년 선수별 3년 평균 안타수

- 선수별 안타 평균 구하기
- 구한 결과를 sort_values를 이용해 정렬까지(안타가 많을 수록 높은 순위)

```python
hit_result = nc_result1.groupby('선수명')[['안타']].mean().sort_values(by='안타',ascending=False)
hit_result.head()
```

```python
hit_player=hit_result.head(10).index.tolist()
hit_player
```

## 평균 안타수 기준 상위 10위 타자들의 3년간 안타, 홈런, 타율 조회

- 참고) NaN값이 있을 경우 '-'로 대체해서 표현

```python
# 조건 색인시 .isin(리스트)를 사용하면, 해당 리스트에 포함된 사람만 조회 가능
# fancy인덱싱(조건 색인에서 미리 준비된 리스트에 해당하는 목록만 남김)
top10_player=nc_result1[nc_result1['선수명'].isin(hit_player)]
top10_player.head()
```

```python
top10_player=top10_player[['연도','선수명','안타','홈런','타율']]
top10_player
```

```python
# 연도, 선수명을 로우로 내리기
# df.set_index(['내릴컬럼들'])
top10_player=top10_player.set_index(['연도','선수명'])
top10_player
```

```python
# 위 자료에 대해 unstck으로 "연도"를 컬럼으로 올리기
top10_player.unstack(0)
```

```python
# NaN를 다른 데이터로 교체하기: .fillna('채울자료')
top10_player.unstack(0).fillna('-')
```

## 연도별 팀의 총 안타수, 총 홈런수, 총 도로수

```python
team_result=nc_result1.groupby('연도')[['안타','홈런','도루']].sum()
team_result
```

```python
# 엑셀이나 csv로 빼는 방법
# 엑셀: to_excel("경로 및 파일명")
# csv: to_csv("")
team_result.to_excel("NC result.xlsx")
```
