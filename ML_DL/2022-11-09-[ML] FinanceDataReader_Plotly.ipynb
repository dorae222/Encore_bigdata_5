```python
import FinanceDataReader as fdr

# 삼성전자(005930) 전체 (1996-11-05 ~ 현재)
df = fdr.DataReader('005930')
df
```

```python
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
fig = go.Figure([
    go.Scatter(name="Open",x=df.index,y=df['Open'],mode='lines',line=dict(color='#ffeead',width=2)),
    go.Scatter(name="High",x=df.index,y=df['High'],mode='lines',line=dict(color='#96ceb4',width=2)),
    go.Scatter(name="Low",x=df.index,y=df['Low'],mode='lines',line=dict(color='#88d8b0',width=2)),
    go.Scatter(name="Close",x=df.index,y=df['Close'],mode='lines',line=dict(color='#ffcc5c',width=2)),
    go.Scatter(name="Volume",x=df.index,y=df['Volume'],mode='lines',line=dict(color='#ff6f69',width = 2)),
    go.Scatter(name="Change",x=df.index,y=df['Change'],mode='lines',line=dict(color='#ff6f69',width = 2))
    ])

# 배경 설정
fig.update_layout(plot_bgcolor='rgb(249,249,249)', margin=dict(l=0, r=0, t=0, b=0))

# 기간 버튼 설정
fig.update_layout(
    title_text='삼성전자 주식 정보',
    title_x=0.5,
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1,
                         label="한달",
                         step="month",
                         stepmode="backward"),
                    dict(count=6,
                         label="6개월",
                         step="month",
                         stepmode="backward"),
                    dict(count=1,
                         label="1년",
                         step="year",
                         stepmode="backward"),
                    dict(count=1,
                         label="올해",
                         step="year",
                         stepmode="todate"),
                    dict(label="전체",
                         step="all")
                ])
            ),
            rangeslider=dict(
                visible=True
            ),
            type="date"
        )
    )


fig.show()
```
